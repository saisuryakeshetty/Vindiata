```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel('Data Science Case study Vindiata.xls','Operations', index_col =0,header = 3,usecols = 'B:O')
```

```python
df1.head()
```

```python
df2 = pd.read_excel('Data Science Case study Vindiata.xls','AC characteristics',index_col = 0 ,header = 3,usecols = 'B:F')
```

```python
df2
```

```python
total_hours_Series = df1.groupby('Aircraft Type').sum().sum(axis = 1)
```

```python
df2['total_hours'] = total_hours_Series
```

```python
df2
```

```python
total_cost_Series = df2[['Costs per flight hour','total_hours']].apply(lambda A: A[0]*A[1],axis = 1)
```

```python
df2['total_cost'] = total_cost_Series
```

```python
df2
```

```python
df2['total_cost']
```

```python
cost_per_seat_per_km_Series = df2[['Ave. Speed (km/h)','Number of Seats','Costs per flight hour']].apply(lambda A: A[2]/(A[0]*A[1]),axis = 1) 
```

```python
df2['cost_per_seat_per_km'] = cost_per_seat_per_km_Series
```

```python
df2
```

```python
df2['cost_per_seat_per_km']
```
