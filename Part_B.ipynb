```python
import numpy as np
import pandas as pd
```

```python
df2 = pd.read_excel('Data Science Case study Vindiata.xls','AC characteristics',header = 3,usecols = 'B:F')
```

```python
df2
```

```python
df3 = pd.read_excel('Data Science Case study Vindiata.xls','City pairs',header = 3,usecols = 'B:E')
```

```python
df3
```

```python
df2['key'] = 1
```

```python
df3['key'] = 1
```

```python
df4 = pd.merge(df3,df2,on=['key'],how ='inner').drop(['key'],axis=1)
```

```python
df4.head(10)
```

```python
df4 = df4[df4['Distance (km)']<=df4['Range (Km)']]
```

```python
Number_of_trips_Series = df4[['Pass. Demand \n(per day)','Number of Seats']].apply(lambda A: np.ceil(A[0]/A[1]),axis = 1)
```

```python
df4['Number_of_trips'] = Number_of_trips_Series 
```

```python
df4.head()
```

```python
cost_per_day_Series = df4[['Distance (km)','Ave. Speed (km/h)','Costs per flight hour','Number_of_trips']].apply(lambda A: A[0]*A[2]*A[3]/A[1],axis =1)
```

```python
df4['cost_per_day'] = cost_per_day_Series
```

```python
df4.head()
```

```python
df4.groupby(['Origin City','Desitnation City'])['cost_per_day'].min()
```

```python
pd.merge(df4,df4.groupby(['Origin City','Desitnation City'])['cost_per_day'].min(),on='cost_per_day',how ='inner')
```
